```python
# Dependencies
import tweepy
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
import seaborn as sns
import spacy
from config import consumer_key, consumer_secret, access_token, access_token_secret


# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Authorization    
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_md')
```

```python
# Function to get tweet text, created_at, retweets and adds a column of datetime objects for tweet time

def get_tweets(user, num=200):
    # Take a twitter id and number of tweets desired, returns a dataframe of tweet texts, created_at, and datetime objects for tweet time
    # Get tweet text and created_at objects
    # Put into dataframe
    tweets = pd.DataFrame([status._json for status in tweepy.Cursor(api.user_timeline, id=user, tweet_mode="extended")
             .items(num)])[["full_text", "created_at", "retweet_count"]]
    
    # Add a column of datetime objects for each tweet
    tweets["datetime_object"] = tweets.created_at.map(lambda x: datetime.strptime(x, "%a %b %d %H:%M:%S +0000 %Y"))
    return tweets
```

```python
# Function to calculate the time differences between each tweet

def calculate_time_diffs(list_from_df):
    # Takes a list of datetime objects and returns a list of time differences
    time_deltas = list(zip(list_from_df[0:], list_from_df[1:]))

    # Calculate the time difference between each deltas
    # Account for the possibility of following tweet being sent on a following day
    time_differences = [((time[0] - time[1]).seconds / 3600) + ((time[0] - time[1]).days) for time in time_deltas]
    return time_differences
```

```python
# Function to plot tweet velocity

def tweet_velocity(name, time_diffs, color="r"):
    # Takes twitter user name (string), a list of time differences, color (string) and returns plot of tweet velocity.
    
    plt.plot(range(len(time_diffs)), time_diffs, marker="o", color=color, alpha=0.25)
    plt.title(f"Tweet Velocity of {name}")
    plt.xlabel("Tweets Ago")
    plt.ylabel("Hours Ago")
    plt.figure(figsize=(20,10))
    
    return plt.show()
```

```python
# Getting 300 tweets

trump_tweets = get_tweets("realDonaldTrump", 300)
```

```python
# Creating a column of created_times as an int, format year, month, day, hour, minute

trump_tweets["int_time"] = trump_tweets.datetime_object.map(lambda x: int(datetime.strftime(x, "%Y%m%d%H%M")))
trump_tweets.head()
```

```python
# Getting a weekago as an int
now = datetime.now()
print(type(now))
print(now)

weekago = int(datetime.strftime(now - timedelta(days=7), "%Y%m%d%H%M"))
print(type(weekago))
print(weekago)
```

```python
# Selecting only the rows that are greater than a weekago

trump_week = trump_tweets.loc[trump_tweets["int_time"] >= weekago]
trump_week
```

```python
trump_time_diffs = calculate_time_diffs(trump_week.datetime_object)
weekly_velocity = tweet_velocity("realDonaldTrump", trump_time_diffs, "red")
```

```python
# Grouping by hour and day
print(len(trump_week))
trump_week.tail()
```

```python
# Add columns for hour and weekday created
trump_week["hour"] = trump_week.datetime_object.map(lambda x: x.strftime("%H %p"))
trump_week["weekday"] = trump_week.datetime_object.map(lambda x: x.strftime("%a"))
trump_week.head()
```

```python
# Groupby hour and weekday columns
trump_week_grouped =trump_week.groupby(["hour", "weekday"]).count().unstack("weekday", fill_value=0)["full_text"]

# Reorder the days of the week
trump_week_grouped = trump_week_grouped[["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]]
trump_week_grouped
```

```python
# Heatmap gradient

cm = sns.light_palette("red", as_cmap=True)

trump_times_heatmap = trump_week_grouped.style.background_gradient(cmap=cm)
trump_times_heatmap
```

```python
# Analyze each the polarity of each tweet

trump_week["polarity_compound_scores"] = trump_week.full_text.map(lambda x: analyzer.polarity_scores(x)["compound"])
trump_week.head()
```

```python
# Create another dataframe with hour, day, and scores


grouped_trump_scores = trump_week[["hour", "weekday", "polarity_compound_scores"]]
grouped_trump_scores = grouped_trump_scores.groupby(["hour", "weekday"]).mean().unstack("weekday", fill_value=0)["polarity_compound_scores"]
grouped_trump_scores = grouped_trump_scores[["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]]
grouped_trump_scores
```

```python
# Heatmap

plt.figure(figsize=(20, 10))
midpoint = 0
grouped_trump_scores = sns.heatmap(grouped_trump_scores, cmap="RdBu_r", robust=True, annot=True, center=midpoint)
plt.title("Trump Polarity Scores")
grouped_trump_scores.xaxis.tick_top()
grouped_trump_scores.set_xlabel("Weekday")
grouped_trump_scores.xaxis.set_label_position('top') 
grouped_trump_scores.set_ylabel("Hour")
```

```python
# Functions for natural language processing entity text and entity labels

def entity_text(string):
    doc = nlp(string)
    texts = [ent.text for ent in doc.ents]
    return texts
```

```python
# Adding nlp texts to the dataframe

trump_week["nlp_text"] = trump_week.full_text.map(entity_text)
```

```python
# Counting the texts in each list

# Dictionary to hold count
trump_text_count = {}

# Loop through each line of the dataframe
for lst in trump_week["nlp_text"]:
    for item in lst:
        if item not in trump_text_count:
            text_entry = {item: +1}
            trump_text_count.update(text_entry)
        else:
            trump_text_count[item] += 1
            

trump_text_count
```

```python
# Dataframe of the count

counted_texts = pd.Series(trump_text_count).to_frame(name="count")
counted_texts
```

```python
# Ordering the texts

counted_texts.sort_values(by="count", ascending=False, inplace=True)
counted_texts
```

```python
# Bar plot of most tweeted about nlp texts

counted_texts[:20].plot.bar(subplots=True, color="red", figsize=(20,10))
plt.title("Top Twenty Subjects of realDonaldTrump Tweets")
```
